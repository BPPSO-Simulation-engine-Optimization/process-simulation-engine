# Simulation Demo: Next-Activity Workflow

Minimal walkthrough that reuses the advanced next-activity models to drive a simple case simulation (no resource allocation).


```python
# 1) Imports (add project root so advanced/ is on sys.path)
import os, sys
sys.path.insert(0, os.path.dirname(os.path.dirname(os.getcwd())))  # adds process-simulation-engine/

import pandas as pd
from advanced import (
    load_models,
    load_simulation_assets,
    simulate_cases_advanced,
    events_to_dataframe,
)
```

```python
# 2) Load models and assets
MODELS_DIR = os.path.join(os.getcwd(), "..", "models_lstm_new")
ASSETS_DIR = os.path.join(os.getcwd(), "..", "assets")

models = load_models(MODELS_DIR)
process_graph, decision_map = load_simulation_assets(
    transitions_path=os.path.join(ASSETS_DIR, "named_transitions.json"),
    decision_map_path=os.path.join(ASSETS_DIR, "bpmn_decision_point_map.pkl"),
)

print(f"Loaded models: {len(models)}")
print(f"Process nodes: {len(process_graph)} | Decisions: {len(decision_map)}")
```

```python
# 3) Sample cases
cases = [
    {"loan_goal": "Car", "application_type": "New credit"},
    {"loan_goal": "Home improvement", "application_type": "Limit raise"},
]

sim_events = simulate_cases_advanced(
    cases,
    models=models,
    process_graph=process_graph,
    decision_point_map=decision_map,
)

len(sim_events)
```

```python
# 4) Inspect events
sim_df = events_to_dataframe(sim_events)
sim_df.head()
```
